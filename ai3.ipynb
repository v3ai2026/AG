### Subtask: Implement `pushProjectToGithub` in `services/githubService.ts`

To implement the `pushProjectToGithub` method in `services/githubService.ts`, you will need to:

1.  **Locate the file**: Open the `services/githubService.ts` file in your project.
2.  **Add the following asynchronous function**:

    ```typescript
    import { Octokit } from '@octokit/core';

    // You might need to install Octokit: npm install @octokit/core
    // Or if you're using a different HTTP client, adjust accordingly.

    export async function pushProjectToGithub(
      githubToken: string,
      projectName: string,
      files: { [path: string]: string }
    ): Promise<string> {
      const octokit = new Octokit({ auth: githubToken });
      const owner = 'YOUR_GITHUB_USERNAME'; // Replace with the actual GitHub username or get it from context

      try {
        // 1. Check if repository exists
        let repoExists = false;
        try {
          await octokit.request('GET /repos/{owner}/{repo}', {
            owner,
            repo: projectName,
          });
          repoExists = true;
        } catch (error: any) {
          if (error.status === 404) {
            console.log(`Repository '${projectName}' does not exist. Creating a new one.`);
          } else {
            throw error; // Re-throw other errors
          }
        }

        // 2. Create repository if it doesn't exist
        if (!repoExists) {
          await octokit.request('POST /user/repos', {
            name: projectName,
            private: false, // Or true, depending on requirements
            description: `Generated project for ${projectName}`,
          });
          console.log(`Repository '${projectName}' created successfully.`);

          // 3. Initialize with a README.md if new
          const readmeContent = `# ${projectName}\n\nThis project was generated automatically.`;
          await octokit.request('PUT /repos/{owner}/{repo}/contents/{path}', {
            owner,
            repo: projectName,
            path: 'README.md',
            message: 'Initialize repository with README.md',
            content: Buffer.from(readmeContent).toString('base64'),
          });
          console.log('README.md initialized.');
        }

        // 4. Push all generated files
        for (const filePath in files) {
          if (Object.prototype.hasOwnProperty.call(files, filePath)) {
            const fileContent = files[filePath];
            const commitMessage = `Add/Update ${filePath}`;

            // Check if file exists to determine if it's an update or create
            let sha = undefined;
            try {
              const { data } = await octokit.request('GET /repos/{owner}/{repo}/contents/{path}', {
                owner,
                repo: projectName,
                path: filePath,
              });
              if ('sha' in data) {
                sha = data.sha;
              }
            } catch (error: any) {
              if (error.status !== 404) {
                console.warn(`Could not get SHA for ${filePath}:`, error.message);
              }
            }

            await octokit.request('PUT /repos/{owner}/{repo}/contents/{path}', {
              owner,
              repo: projectName,
              path: filePath,
              message: commitMessage,
              content: Buffer.from(fileContent).toString('base64'),
              sha: sha, // Include SHA for updates
            });
            console.log(`File '${filePath}' pushed to '${projectName}'.`);
          }
        }

        return `Successfully pushed project '${projectName}' to GitHub.`;
      } catch (error: any) {
        console.error('Error pushing project to GitHub:', error);
        // Handle specific errors like authentication failures (401), rate limits (403)
        if (error.status === 401) {
          throw new Error('GitHub authentication failed. Check your token.');
        } else if (error.status === 403) {
          throw new Error('GitHub API rate limit exceeded or forbidden. Try again later.');
        } else {
          throw new Error(`Failed to push project to GitHub: ${error.message}`);
        }
      }
    }
    ```

### Explanation of the steps:

*   **Import `Octokit`**: This is a GitHub API client library. You'll likely need to install it (`npm install @octokit/core`).
*   **Initialize `Octokit`**: The `githubToken` is used to authenticate with the GitHub API.
*   **Check Repository Existence**: A `GET` request to `/repos/{owner}/{repo}` is made. If it returns a 404 status, the repository doesn't exist.
*   **Create Repository**: If the repository doesn't exist, a `POST` request to `/user/repos` creates a new public repository.
*   **Initialize with `README.md`**: For a newly created repository, a `PUT` request to `/repos/{owner}/{repo}/contents/{path}` is used to add a basic `README.md` file. The content must be Base64 encoded.
*   **Push Generated Files**: The function iterates through the `files` object. For each file, it first attempts to get its SHA to determine if it's an update. Then, it uses a `PUT` request to `/repos/{owner}/{repo}/contents/{path}` to create or update the file. The file content is Base64 encoded, and a commit message is provided.

## 实现 GitHub 服务方法

### Subtask:
在 `services/githubService.ts` 中实现一个新方法，该方法将接受 GitHub token、项目名称和生成的文件内容作为参数。此方法将负责创建仓库（如果不存在）、使用一个基本的 `README.md` 文件初始化它，然后将所有生成的文件推送到该仓库。


**Reasoning**:
The subtask requires creating the `services/githubService.ts` file. Since it doesn't exist, I first need to create the parent `services` directory. Then, I will write the TypeScript code for the `pushProjectToGithub` function into this file.



````python
import os

# Create the services directory if it doesn't exist
os.makedirs('services', exist_ok=True)

file_content = '''import { Octokit } from '@octokit/core';

// You might need to install Octokit: npm install @octokit/core
// Or if you're using a different HTTP client, adjust accordingly.

export async function pushProjectToGithub(
  githubToken: string,
  projectName: string,
  files: { [path: string]: string }
): Promise<string> {
  const octokit = new Octokit({ auth: githubToken });
  const owner = 'YOUR_GITHUB_USERNAME'; // Replace with the actual GitHub username or get it from context

  try {
    // 1. Check if repository exists
    let repoExists = false;
    try {
      await octokit.request('GET /repos/{owner}/{repo}', {
        owner,
        repo: projectName,
      });
      repoExists = true;
    } catch (error: any) {
      if (error.status === 404) {
        console.log(`Repository '${projectName}' does not exist. Creating a new one.`);
      } else {
        throw error; // Re-throw other errors
      }
    }

    // 2. Create repository if it doesn't exist
    if (!repoExists) {
      await octokit.request('POST /user/repos', {
        name: projectName,
        private: false, // Or true, depending on requirements
        description: `Generated project for ${projectName}`,
      });
      console.log(`Repository '${projectName}' created successfully.`);

      // 3. Initialize with a README.md if new
      const readmeContent = `# ${projectName}\n\nThis project was generated automatically.`;
      await octokit.request('PUT /repos/{owner}/{repo}/contents/{path}', {
        owner,
        repo: projectName,
        path: 'README.md',
        message: 'Initialize repository with README.md',
        content: Buffer.from(readmeContent).toString('base64'),
      });
      console.log('README.md initialized.');
    }

    // 4. Push all generated files
    for (const filePath in files) {
      if (Object.prototype.hasOwnProperty.call(files, filePath)) {
        const fileContent = files[filePath];
        const commitMessage = `Add/Update ${filePath}`;

        // Check if file exists to determine if it's an update or create
        let sha = undefined;
        try {
          const { data } = await octokit.request('GET /repos/{owner}/{repo}/contents/{path}', {
            owner,
            repo: projectName,
            path: filePath,
          });
          if ('sha' in data) {
            sha = data.sha;
          }
        } catch (error: any) {
          if (error.status !== 404) {
            console.warn(`Could not get SHA for ${filePath}:`, error.message);
          }
        }

        await octokit.request('PUT /repos/{owner}/{repo}/contents/{path}', {
          owner,
          repo: projectName,
          path: filePath,
          message: commitMessage,
          content: Buffer.from(fileContent).toString('base64'),
          sha: sha, // Include SHA for updates
        });
        console.log(`File '${filePath}' pushed to '${projectName}'.`);
      }
    }

    return `Successfully pushed project '${projectName}' to GitHub.`;
  } catch (error: any) {
    console.error('Error pushing project to GitHub:', error);
    // Handle specific errors like authentication failures (401), rate limits (403)
    if (error.status === 401) {
      throw new Error('GitHub authentication failed. Check your token.');
    } else if (error.status === 403) {
      throw new Error('GitHub API rate limit exceeded or forbidden. Try again later.');
    } else {
      throw new Error(`Failed to push project to GitHub: ${error.message}`);
    }
  }
}'''

with open('services/githubService.ts', 'w') as f:
    f.write(file_content)

print("Created services/githubService.ts successfully.")
````
